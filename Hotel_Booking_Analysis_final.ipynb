# **Project Summary -**

The hotel booking analysis project aimed to analyze a dataset containing information about hotel bookings for City and Resort hotels. The dataset contained 119,390 rows and 32 columns, and the analysis was divided into three main categories: data collection, data cleaning and manipulation, and exploratory data analysis (EDA).

The data collection phase involved gathering information about the different columns in the dataset using various methods such as head(), tail(), info(), describe(), and columns(). The columns of interest were hotel, is_canceled, lead_time, arrival_date_year, arrival_date_month, arrival_date_week_number, arrival_date_day_of_month, stays_in_weekend_nights, stays_in_week_nights, adults, children, babies, meal, country, market_segment, distribution_channel, is_repeated_guest, previous_cancellations, previous_bookings_not_canceled, reserved_room_type, assigned_room_type, booking_changes, deposit_type, agent, company, days_in_waiting_list, customer_type, adr, required_car_parking_spaces, total_of_special_requests, reservation_status, reservation_status_date, total_stay, and total_guests.

During the data cleaning and manipulation phase, duplicates were identified and removed, null values were filled, and some columns were dropped due to a high number of null values. The EDA phase involved using various charts and statistical methods to draw insights from the data. The findings from the analysis are discussed below.

The first observation was that the majority of bookings were made for City Hotels compared to Resort Hotels. The Resort hotel had a higher number of repeated guests than the City hotel. However, the City hotel had a higher number of canceled bookings, indicating that the City hotel had more customer turnover than the Resort hotel.

The preferred room type among guests was room type A, followed by room type D. The most popular months for booking a hotel were July and August, indicating that these months were the peak tourist season.

The lead time for bookings was higher in the City hotel compared to the Resort hotel. Guests tended to stay more on weeknights than weekends, and the most common length of stay was 1-3 nights. The agent with ID '9' made the highest number of bookings, which amounted to 28,721 in total.

The analysis showed that the ADR experiences linear growth from June to August, indicating that these months were the peak season for hotel prices. The most common meal type booked was Bed & Breakfast. The top countries of origin for guests were Portugal, the United Kingdom, and Spain.

Lastly, ADR grew linearly with the total number of guests. The average daily rate for a hotel room was around 100 euros. These findings provide valuable insights to hotel managers and marketers, who can use them to optimize their pricing strategies, marketing efforts, and overall customer experience.
